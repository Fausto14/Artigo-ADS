## PREPARANDO A BASE

```python
import numpy as np
import pandas as pd
from numpy import savetxt
```

### Base original : 24 atributos, CM = 64 n√≠veis de cinza

```python
# Base sem crises
base_original = pd.read_csv('FEATURES_M73_N4.csv', sep= ',', names=['F'+str(i) for i in range(25)])
base_original
```

```python
X = base_original.loc[:, base_original.columns != 'F24'].values
y = base_original[['F24']].values
```

```python
def normalization(X, normalization_type='std'):
    X_mean = X.mean(axis = 0)
    X_std = X.std(axis = 0)
    return (X-X_mean)/ X_std, X_mean, X_std
```

```python
nomes_atributos = [
    'ASM', 'Contrast', 'Correlaction', 'Sum of squares',
    'IDM', 'Sum Average', 'Sum Entropy', 'Sum Variance',
    'Entropy', 'Difference Variance', 'Difference Entropy', 'IMC1',
    'IMC2', 'MCC', 'Homogeneity', 'Sum Mean',
    'Maximum Probability', 'Cluster Tendency', 'Cluster Shade',
    'Cluster Prominence', 'Dissimilarity', 'Difference Mean', 'Autocorrelation', 'Inertia', 'Class']
```

```python
X_norm, X_mean, X_std = normalization(X)
data_norm = np.concatenate((X_norm,y),axis=1)
base_original_norm = pd.DataFrame(data_norm, columns=nomes_atributos)
base_original_norm.rename(columns={"F24": "Class"}, inplace = True)
base_original_norm['Class'] = base_original_norm['Class'].astype(int)
base_original_norm
```

```python
base_original_norm.to_csv(path_or_buf='norm_24_features_m73_n4.csv', sep=',',header= True, index=False)
```

#### Selecionando os 10 melhores atributos

```python
nomes_10_melhores_atributos = ['ASM', 'Contrast', 'IDM', 'Entropy', 'Homogeneity', 'Sum Mean',
       'Maximum Probability', 'Dissimilarity', 'Difference Mean',
       'Autocorrelation', 'Class']
base_10_original_norm = base_original_norm[nomes_10_melhores_atributos]
base_10_original_norm
```

```python
base_10_original_norm.to_csv(path_or_buf='norm_10_features_m73_n4.csv', sep=',',header= True, index=False)
```

#### Classifica√ß√£o Multiclasse para Bin√°ria
<p>
    Classe Superior (melhor qualidade) e Classe Inferior (pior qualidade).
</p>
<p>
    No ambiente real existem 7 classes, sendo que de n√∫mero 1 apresenta melhor qualidade e a de n√∫mero 7 apresenta uma qualidade inferior. 
</p>
<p>
    Para este trabalho o problema foi transformado em bin√°rio da seguinte maneira:
    
- As classes de 1 a 5 foram unidas e passaram a ser denominadas de classe
    <b>Superior (Class: 1)</b>;
- As classes 6 e 7 tamb√©m foram unidades e passaram a ser denominadas de
    classe <b>Inferior (Class: 0)</b>.
</p>

```python
def multiToBin(label):
    if (label == 1 or label == 2 or label == 3 or label == 4 or label == 5):
        return 1;
    else:
        return 0;
```

```python
def multiToBin2(label):
    if (label == 1 or label == 2 or label == 3 or label == 4):
        return 1;
    else:
        return 0;
```

```python
bin_base_10_original_norm = base_10_original_norm.copy()
bin_base_10_original_norm['Class'] = bin_base_10_original_norm.apply(lambda row : multiToBin(row['Class']), axis = 1)
bin_base_10_original_norm['Class'].value_counts()
```

```python
bin_base_10_original_norm2 = base_10_original_norm.copy()
bin_base_10_original_norm2['Class'] = bin_base_10_original_norm2.apply(lambda row : multiToBin2(row['Class']), axis = 1)
bin_base_10_original_norm2['Class'].value_counts()
```

```python
bin_base_10_original_norm.to_csv(path_or_buf='bin_norm_10_features_m73_n4.csv', sep=',',header= True, index=False)
bin_base_10_original_norm2.to_csv(path_or_buf='2bin_norm_10_features_m73_n4.csv', sep=',',header= True, index=False)
```

### T-Distributed Stochastic Neighbouring Entities (t-SNE)

```python
X = base_10_original_norm.loc[:, base_10_original_norm.columns != 'Class'].values
y = base_10_original_norm['Class'].values
base_10_original_norm['Class'].value_counts()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
```

```python
X_2d = tsne.fit_transform(X)
```

```python
target_ids = range(1,8)
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(figsize=(12, 6))
colors =  ['#ff7f0e', '#1f77b4', '#008000', '#FFFF00', '#EE82EE', '#00FFFF' , '#C0C0C0']
for i, c, label in zip(target_ids, colors, ['Classe '+str(i+1) for i in range(7)]):
    plt.scatter(X_2d[y == i, 0], X_2d[y == i, 1], c=c, label=label)

plt.title("t-SNE com compoentes PCA")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.legend()
plt.savefig('tsne_multi.png')
```

```python
X_bin = bin_base_10_original_norm.loc[:, bin_base_10_original_norm2.columns != 'Class'].values
y_bin = bin_base_10_original_norm['Class'].values
bin_base_10_original_norm['Class'].value_counts()
```

```python
from sklearn.manifold import TSNE
tsne_bin = TSNE(n_components=2, random_state=0)
X_2d_bin = tsne_bin.fit_transform(X_bin)
```

```python
target_ids = list(range(1,3))
target_ids.reverse()
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(figsize=(12, 6))
colors =  ['#ff7f0e','#008000']
for i, c, label in zip(target_ids, colors, ['Classe 0', 'Classe 1']):
    if(i == 1):
        s = np.concatenate((X_2d_bin[y == 1], X_2d_bin[y == 2], X_2d_bin[y == 3], X_2d_bin[y == 4]), axis=0)
    else:
        s = np.concatenate((X_2d_bin[y == 5], X_2d_bin[y == 6], X_2d_bin[y == 7]), axis=0)
        
    plt.scatter(s[:,0], s[:,1], c=c, label=label)

plt.title("t-SNE com compoentes PCA")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.legend()
plt.savefig('tsne_bin.png')
```
